```
'''
GetUrl
住友不動産のページ内のリンクを取得する。
x:aタグからURLを取得したいページのURL
y:取得したいHPのhomeのURL
'''


def GetUrl (x,y):
    headers = {
    'User-Agent':'Mozilla/5.0',
    'referer':y
    }
    r = requests.get(x,headers=headers)
    html = r.text
    soup = BeautifulSoup(html, 'html.parser')
    links = [url.get('href') for url in soup.find_all('a', class_='textLink post_param blank_window')]
    return links
```
